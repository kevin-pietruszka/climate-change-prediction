```python
# from geopy.geocoders import Nominatim
import pandas as pd
import json 
df = pd.read_csv("./data/GlobalLandTemperaturesByCity.csv")
```

```python
# cities = df['City'].unique()

cities = df.groupby('City').groups
print(len(cities))
```

```python
info = [['', '', '', ''] for i in range(len(cities))]

i = 0
for city, idxs in cities.items():
    row = df.iloc[idxs[0]]
    lat = str(row['Latitude'])
    lng = str(row ['Longitude'])
    
    latitude = float(lat[:len(lat) - 1])
    longitude = float(lng[:len(lng) - 1])
    
    if lat[-1] == 'S':
        latitude *= -1
    if lng[-1] == 'W':
        longitude *= -1
    
    info[i] = [city, row['Country'], latitude, longitude] 
    
    i += 1
    
output_df = pd.DataFrame(info, index=None ,columns=['City', 'Country', "Latitude", 'Longitude'])

output_df.to_csv('./data/city_locations.csv')

output_df
```

```python
with open("./visualization/cities.json", "w") as f:
    json.dump(info, f)
```
